```python
import os
import math
import matplotlib.pyplot as plt
import numpy as np
from pytlm import Log
```

## Plot one payload from multiple sessions

This is very useful if you know that something happened, but you don't know in which log that happens

```python
path = 'logs/telemetry/endurance/'
for session in os.listdir(path):
    log = Log(os.path.join(path, session), resample=True, resample_interval='1ms', consider_networks=['inverters'], align='beginning')

    print(session)
    log.quick_plot('inverters', 'inv_r_id_4a_t_igbt', 't_igbt')
```

## Plot everything from a session

Easily access all the logged data of a session

#### Load log

```python
log = Log('logs/telemetry/endurance/acceleration_[]_003/', resample=True, resample_interval='1ms', consider_networks=['inverters'], align='beginning')
```

#### Plot signals

```python
plt.rcParams["figure.autolayout"] = True

for network in log.get_networks_list():
    for message in log.get_messages_list(network):
        for payload in log.get_payloads_list(network, message):
            data = log.get(network, message, payload)
            if data.isin([0, 1]).all():
                # ignore binary flags
                continue

            log.quick_plot(network, message, payload)
```
